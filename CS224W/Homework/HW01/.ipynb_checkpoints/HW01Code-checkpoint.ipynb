```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

# Problem 2

```python
G = nx.Graph()
G.add_nodes_from([(i+1, {"weight": 0.5}) for i in range(10)])
G.add_edges_from([(1, 2), (1, 3), (2, 3), (2,4), (2,5), (3,6), (4, 5), (4, 7), (4, 8), (5, 6), (5, 9), (6, 10), (8, 9)])
G.nodes[5]["weight"] = 0
G.nodes[6]["weight"] = 1
G.nodes[7]["weight"] = 1
G.nodes[10]["weight"] = 1


nx.draw(G, with_labels = True)
```
