# Compute pi by finding the area of a circle

```python
from numpy.random import uniform
import matplotlib.pyplot as plt
import numpy as np
N = 200000
pts = uniform(-1, 1, (N,2))
dist = np.linalg.norm(pts, axis =1 )
in_circle = dist <= 1

pts_in_circle = np.count_nonzero(in_circle)
pi = 4 * pts_in_circle/N

plt.scatter(pts[in_circle, 0], pts[in_circle, 1], marker=',', edgecolors='k', s=1)
plt.scatter(pts[~in_circle, 0], pts[~in_circle, 1], marker=',', edgecolors='r', s=1)
plt.axis('equal')
print(pi)
```
