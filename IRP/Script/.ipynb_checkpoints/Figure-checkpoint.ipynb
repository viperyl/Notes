```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
```

# Six basic emotion

```python
path = "/run/media/viper/LSP/Dataset/FER2013/train"
emotions = ['angry', 'disgust', 'fear', 'happy', 'sad', 'surprise']
```

```python
fig, ax = plt.subplots(1, 6, figsize=(16,20))
for i in range(6):
    files_path = os.path.join(path, emotions[i])
    index = np.random.randint(0, len(os.listdir(files_path)))
    file = os.listdir(files_path)[index]
    img = cv2.imread(os.path.join(path, emotions[i], file))
    ax[i].imshow(img)
    ax[i].set_axis_off()
    ax[i].set_title(emotions[i], fontsize=20)
plt.savefig('six_emotions.eps', format='eps', bbox_inches='tight')
```
